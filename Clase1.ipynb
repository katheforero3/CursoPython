**Operadores Matematicos**

```python
a=2
b=4

#Suma
print(a+b) #El resultado es 6

#Resta
print(b-a) #El resultado es 2

#División
c=b/a #El resultado es 2
print(c)

#División entera
d=3
print(d//a) #El resultado es 1

#Exponenciación
print(a**5) #El resultado es 32

#Modulo
print(b%a) #El resultado es 0

#Asignación compuesta
a+=b
print(a) #El resultado es 6
a=2

#Redondeo
e=3.14159
print(round(e,2)) #El resultado es 3.14

#Valor absoluto
f=a-b
print(abs(f)) #El resultado es 2
```

**Valores String**

```python
#Concatenación de caracteres
saludo='Hola'
nombre='Mundo'
mensaje=saludo+" , "+nombre
print(mensaje)
```

```python
#Repetición de caracteres
repetido=saludo*3
print(repetido)
```

```python
#Indexación
#Obtener una variable que tenga el primer caracter de la variable saludo
primer_caracter=saludo[0]
print(primer_caracter)

#Genere nueva variable donde extraiga la palabra hola
subcadena=mensaje[0:4]
print(subcadena)
```

```python
#Métodos de cadenas

#Mayusculas
cadena='Python es genial'
print(cadena.upper())

#Minusculas
print(cadena.lower())

#Reemplazo
print(cadena.replace('genial','ingreible'))

#Separar
print(cadena.split(' '))

print(cadena[0])

cadena_separado=cadena.split(' ')
print(cadena_separado[0])
```
